###### Department wise codekata performence (pie chart)

```python
performance = df.groupby('Department').sum()['CodeKata Score']
performance
```

```python
data = df.groupby('Department').sum()['CodeKata Score']
keys = df['Department'].unique()
myexplode = [0.1, 0, 0]
data
palette_color = seaborn.color_palette('dark')
plt.pie(data,labels=keys,colors=palette_color,explode = myexplode, autopct='%.0f%%')
plt.title("Department wise codekata performence (pie chart)", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
```

###### Department wise toppers (horizantal bar graph or any visual representations of your choice)

```python
barWidth = 0.20
fig = plt.subplots(figsize =(10, 6))
CSE_toppers = df[df['Department'] == 'Computer Science and Engineering'].sort_values(by='CodeKata Score',ascending=False).head(3)
ECE_toppers = df[df['Department'] == 'Electronics and Communication Engineering'].sort_values(by='CodeKata Score',ascending=False).head(3)
EEE_toppers = df[df['Department'] == 'Electronics and Electrical Engineering'].sort_values(by='CodeKata Score',ascending=False).head(3)
CSE = CSE_toppers['CodeKata Score']
ECE = ECE_toppers['CodeKata Score']
EEE = EEE_toppers['CodeKata Score']
br1 = np.arange(len(CSE))
br2 = [x + barWidth for x in br1]
br3 = [x + barWidth for x in br2]
plt.bar(br1, CSE, color ='r', width = barWidth,
        edgecolor ='grey', label ='CSE')
plt.bar(br2, ECE, color ='g', width = barWidth,
        edgecolor ='grey', label ='ECE')
plt.bar(br3, EEE, color ='b', width = barWidth,
        edgecolor ='grey', label ='EEE')
plt.xlabel('Ranks', fontweight ='bold', fontsize = 15)
plt.ylabel('CodeKata Score', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(CSE))],
        ['Rank 1', 'Rank 2', 'Rank 3'])
plt.legend()
plt.title("Department wise toppers", bbox={'facecolor':'0.8', 'pad':5})
plt.show()
#the below fig compares the codekata scores of each department toppers(RANK 1,2,3)
```
